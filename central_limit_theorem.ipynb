```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_plot(axes,samples):

    mean = np.mean(samples)
    variance = np.var(samples)

    sns.kdeplot(samples, ax=axes, fill=True, color='skyblue')

    axes.axvline(mean, color='red',linestyle='--',label = f'Mean = {mean:.2f}')
    axes.axvline(mean+np.sqrt(variance),color='green',linestyle=':',label = f'+1 Std Dev')
    axes.axvline(mean-np.sqrt(variance),color='green',linestyle=':',label = f'-1 Std Dev')

    axes.legend()
    
    axes.grid(True)
    
```

```python
data = np.random.randint(0, 1001, size=500)
mean = np.mean(data)
variance = np.var(data)
mean,variance
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
sample1 = np.random.choice(data, size=5, replace=False)
get_plot(axes[0,0],sample1)
axes[0,0].set_title("Sample Distribution with 5 samples")
sample2 = np.random.choice(data, size=10, replace=False)
get_plot(axes[0,1],sample2)
axes[0,1].set_title("Sample Distribution with 10 samples")
sample3 = np.random.choice(data, size=15, replace=False)
get_plot(axes[1,0],sample3)
axes[1,0].set_title("Sample Distribution with 15 samples")
sample4 = np.random.choice(data, size=30, replace=False)
get_plot(axes[1,1],sample4)
axes[1,1].set_title("Sample Distribution with 30 samples")
plt.tight_layout()
plt.show()
```
